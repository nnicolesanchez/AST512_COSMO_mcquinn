# Example ipython notebook script for computations

```python
import numpy as np
import pylab as plt
from sympy import *
from scipy import integrate
%matplotlib inline
```

```python
#cosmological parameters
h = 0.7
OmegaM=0.3
OmegaL = 1- OmegaM  #assumes flat


lightspeed = 3e5  #km/s

#cosmology funcitons

#Hubble function
def Hz(z):
    return 100*h*(OmegaM*(1+z)**3+OmegaL)**.5

#conformal distance
def Dconf(z):
    return lightspeed*integrate.romberg(dConfDistdz, 0, z)

def dConfDistdz(z):
    return 1./Hz(z)

#angular diameter distance
def DA(z):
    return Dconf(z)/(1+z)
```

```python
zarr = np.linspace(0.1,10, num=100)
DAarr = [DA(zarr[i]) for i in range(100)]
plt.plot(zarr, DAarr) 
plt.title('Angular diameter distance');
plt.xlabel('redshift')
plt.ylabel('Mpc')
```
